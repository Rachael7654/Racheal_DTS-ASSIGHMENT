```python


# Function to compute average
def compute_average(scores):
    return sum(scores) / len(scores)

# Function to check pass or fail
def check_pass_fail(average):
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Apply to 5 students
students = {
    "Student1": [70, 65, 80],
    "Student2": [45, 50, 40],
    "Student3": [90, 85, 88],
    "Student4": [30, 35, 25],
    "Student5": [60, 55, 58]
}

# Print results
for name, scores in students.items():
    avg = compute_average(scores)
    result = check_pass_fail(avg)
    print(f"{name} - Average: {avg:.2f} - {result}")

```
