```python
import matplotlib.pyplot as plt


students = ['John','paul','Damian','manasseh','James','caleb','princess','reacheal','Jay','jerry']
English_scores =[78,90,98,78,69,54,78,91,59,55]
maths_scores = [67,67,86,46,38,98,69,84,90,91]
physics_scores = [78,49,76,87,87,56,54,39,90,81]


plt.figure()
plt.plot(students, physics_scores)
plt.title('physics scores - line chart')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.show()

plt.figure()
plt.bar(students, maths_scores)
plt.title('maths scores - bar chart')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.show()

plt.figure()
plt.pie(English_scores, labels=students,autopct='%1.2f%%')
plt.title('English scores - pie chart')
plt.show()



# interpretations Between line_chart,Bar_chart and pie_chart
# THE LINE_CHART:in the physics line_chart it is represented by trains therefore the trains droped at mathew's spot which was score 49 after droping from john's spot which has a score of 78 and
# also droped at marry's spot with the score of 39
# and also raise again up in Jay's spot with the score of 90 which is the highest train limit showened on the line_chart


# BAR CHARTS: The bar chart is represented with bars in the diagram above
#the diagram shows the hist score represented with a bar while the lowest represented with a bar as well but in a shorter form.
# in seeing the diagram you can tell that the lowest score is 38 and the highest score is 98 both represented with a bar.
#also the 38 is on james while the highest 98 is on caleb


# PIE CHART: the pie chart is devided in sections with the scores and names by the side of the pie
# there are calculated and represented in persentage not the actual score as viewed from the scores given above
# here the one with the highest score persentage is 13.07% which is dave as seen in the pie
# while the lowest is caleb with the score persentage of 7.20%
```
